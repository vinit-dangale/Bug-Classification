```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bug_data1.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer_ = CountVectorizer()
```

```python
vectorizer_.fit(df['description'])
```

```python
vectorizer_.vocabulary_
```

```python
vector = vectorizer_.transform(df['description'])
```

```python
vector.toarray()
```

```python
x = vector.toarray()
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred1 = lr.predict(x_test)
```

```python
np.mean(pred1 != y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred1))
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
pred2 = svm.predict(x_test)
```

```python
np.mean(pred2!=y_test)
```

```python
print(classification_report(y_test,pred2))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
pred3 = dtc.predict(x_test)
```

```python
np.mean(pred3!=y_test)
```

```python
print(classification_report(y_test,pred3))
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters=2)
```

```python
cluster.fit(x_train,y_train)
```

```python
pred4 = cluster.predict(x_test)
```

```python
np.mean(pred4 != y_test)
```

```python
print(classification_report(y_test,pred4))
```
